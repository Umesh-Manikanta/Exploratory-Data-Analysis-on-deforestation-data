<h2 style="color:#FEFFAC">Finding In which year which State is having maximum Deforestation : :</h2>


```python
L1=[]
for z in col[8:]:
    d=[]
    for x in states:
        t=dat.get_group(x)[z].sum()/dat.get_group(x).shape[0]
        # print(t)
        d.append(t)
    # print(d)
    L1.append(d)
print(L1)
print(len(L1),len(L1[0]))
```

```python
g=[]
max_year_value=[]
for x in L1:
    i=np.array(x).argmax()
    max_year_value.append(x[i])
    g.append(i)
print(g)
    
```

```python
year_state=states[g]
year_state
from scipy import stats
max_freq_state=stats.mode(g)[0]
print(states[max_freq_state])
```

```python
plt.figure(figsize=(25,9))
fig, ax = plt.subplots()
for i in range(len(y)):
    plt.text( max_year_value[i]+10,i, str(year_state[i]))
ax = sns.barplot(ax = ax, data = data , y = y, x = max_year_value)
ax.set(ylabel='Year',xlabel='Area in Deforestation', title='Year Vs Maximum Deforestation Occured')
```

<h2 style="color:#FEFFAC">Finding In which year which District is having maximum Deforestation : :</h2>


```python
datt=data.groupby(data['subnational2'])
districts=data["subnational2"].unique()
L2=[]
for z in col[8:]:
    d=[]
    for x in districts:
        t=datt.get_group(x)[z].sum()/datt.get_group(x).shape[0]
        # print(t)
        d.append(t)
    # print(d)
    L2.append(d)
print(L2)
print(len(L2),len(L2[0]))
```

```python
g=[]
max_districts_value=[]
for x in L2:
    i=np.array(x).argmax()
    max_districts_value.append(x[i])
    g.append(i)
print(g)
```

```python
year_dist=districts[g]
year_dist
from scipy import stats
max_freq_state=stats.mode(g)[0]
print(districts[max_freq_state])
```

```python
plt.figure(figsize=(25,9))
fig, ax = plt.subplots()
for i in range(len(y)):
    plt.text( max_districts_value[i]+10,i, str(year_dist[i]))
ax = sns.barplot(ax = ax, data = data , y = y, x = max_districts_value)
ax.set(ylabel='Year',xlabel='Area in Deforestation', title='Year Vs Maximum Deforestation Occured')
```
